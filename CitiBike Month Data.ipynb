```python
import pandas as pd
import glob
```

```python
%ls
```

```python
citi = glob.glob('*.csv')
```

```python
citi[:3]
```

```python
frame = pd.DataFrame()
```

```python
list_ = []
```

```python
for file_ in citi:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
```

```python
frame = pd.concat(list_)
```

```python
frame
```

```python
df.to_csv('citi_all.csv')
```

## Trip Duration By Birth Year

```python
tripduration = frame.groupby(['birth year'], as_index=False)['tripduration'].mean()
# tripduration = data.groupby('birth year')['tripduration'].mean()
tripduration.head()
```

```python
tripduration.to_csv('avgtripduration.csv')
```

## Top Station

```python
topstation = frame.groupby(['end station id']).size().reset_index(name='counts')
topstation.sort_values('counts', ascending=False).head(10)
```

## Gender Count

```python
gendertype = frame['gender']
gendertype.head()
```

```python
male_count = 0
female_count = 0

for type in gendertype:
    if type == 1:
        male_count +=1
    elif type == 2:
        female_count +=1
print(f'male count: {male_count}, female count: {female_count}')
```

## Peak Times Hours

```python
peaktimes = frame.groupby(['starttime']).size().reset_index(name='counts')
peaktimes.sort_values('counts', ascending=False).head()
```

## Start Stations

```python
topstation = frame.groupby(['start station id']).size().reset_index(name='counts')
topstation.sort_values('counts', ascending=False).head(10)
```

```python
bottomstation = frame.groupby(['start station id']).size().reset_index(name='counts')
bottomstation.sort_values('counts', ascending=True).head(10)
```

## Subscriber Customer Count

```python
usertype = frame['usertype']
usertype.head()
```

```python
customer_count = 0
subscriber_count = 0

for type in usertype:
    if type == "Customer":
        customer_count +=1
    elif type == "Subscriber":
        subscriber_count +=1
print(customer_count, subscriber_count)
```
